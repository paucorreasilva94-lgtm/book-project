## WEB SCRAPPING

```python
import requests
from bs4 import BeautifulSoup
import time
import random
import pandas as pd

#to avoid the error
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36"
}

def scrap_goodreads(total_books=500):
    books = []
    page = 1
    base_url = "https://www.goodreads.com/list/show/1.Best_Books_Ever"
    
    while len(books) < total_books:
        print(f"Scraping page {page} (Current total: {len(books)})...")
        try:
            url = f"{base_url}?page={page}"
            response = requests.get(url, headers=HEADERS, timeout=30)
            
            if response.status_code == 200:
                soup = BeautifulSoup(response.content, 'html.parser')
                rows = soup.select('tr[itemtype="http://schema.org/Book"]')
                
                for row in rows:
                    if len(books) >= total_books:
                        break
                    title = row.select_one('.bookTitle').text.strip()
                    author = row.select_one('.authorName').text.strip()
                    
                    books.append({
                        "title": title,
                        "author": author,
                        "description": "N/A",
                        "source": "Scraping"
                    })
                page += 1
            else:
                print(f"Failed page {page}. Status: {response.status_code}. Retrying...")
            
            time.sleep(random.uniform(5, 10))
            
        except Exception as e:
            print(f"Connection error on page {page}: {e}. Retrying in 20s...")
            time.sleep(20) 
            
    return pd.DataFrame(books)

df_scraping = scrap_goodreads(total_books=500)
print(f"Scraping completed. Final count: {len(df_scraping)}")
```

## API from open library

```python
import requests
import pandas as pd

def get_books_from_api(subject="fiction", limit=500):
    books = []
    url = f"https://openlibrary.org/subjects/{subject}.json?limit={limit}"
    
    try:
        response = requests.get(url, timeout=30)
        if response.status_code == 200:
            data = response.json()
            works = data.get('works', [])
            
            for work in works:
                author_name = "Unknown"
                if 'authors' in work and len(work['authors']) > 0:
                    author_name = work['authors'][0]['name']
                
                books.append({
                    "title": work.get('title'),
                    "author": author_name,
                    "description": "N/A", # API endpoint doesn't provide full summary
                    "source": "OpenLibrary API"
                })
        else:
            print(f"Error: {response.status_code}")
            
    except Exception as e:
        print(f"Connection error: {e}")
        
    return pd.DataFrame(books)

df_api = get_books_from_api(subject="fiction", limit=500)
print(f" Data pipeline successful. Collected {len(df_api)} books from Open Library API.")
```

## NLP 

#### Text Cleaning 

```python
import re
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are available
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text_advanced(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

# Apply the cleaning process
df_final['clean_text'] = df_final['clean_text'].apply(clean_text_advanced)
```

#### TF-IDF, vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english', max_features=5000)

# Create the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(df_final['clean_text'])

print(f"Matrix shape: {tfidf_matrix.shape}")
```

#### Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get book recommendations
def get_recommendations(title, cosine_sim=cosine_sim):
    # Check if the title exists
    if title not in df_final['title'].values:
        return "Book title not found in dataset."
        
    # Get the index of the book that matches the title
    idx = df_final[df_final['title'] == title].index[0]

    # Get the pairwise similarity scores of all books with that book
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the books based on the similarity scores (descending order)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 5 most similar books (excluding the book itself)
    sim_scores = sim_scores[1:6]

    # Get the book indices
    book_indices = [i[0] for i in sim_scores]

    # Return the titles of the top 5 most similar books
    return df_final['title'].iloc[book_indices]
```

```python
# Example:
recommendations = get_recommendations('The Hunger Games (The Hunger Games, #1)')
print(recommendations)
```

The system successfully implemented a content-based recommendation engine, achieving high accuracy in literary series, as demonstrated with the test case of 'The Hunger Games'

## Unsupervised Learning 

#### K-Means Clustering

First, we use the Elbow Method to determine how many clusters ($k$) represent your dataset's natural structure

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Calculate inertia for different values of k
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(tfidf_matrix)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method to help choose the best number of clusters
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

#### PCA Visualization

Since the matrix has hundreds or thousands of dimensions (features), PCA allows us to project these into a 2D plane so we can see how the books are grouped.

```python
from sklearn.decomposition import PCA

# Reduce the TF-IDF matrix to 2 dimensions for visualization
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(tfidf_matrix.toarray())

# Create a scatter plot to visualize the clusters
plt.figure(figsize=(10, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.6)
plt.title('Book Clusters Visualization using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```

Clustering analysis using K-Means and dimensionality reduction with PCA revealed that the collection does not split into isolated groups, but rather presents a continuous semantic structure. The PCA visualization shows that most of the titles share a common thematic core, while there are specific niches (represented by the scattered groups) that deviate from this core, validating the model's ability to identify both general trends and unique subgenres.

## Improve Recommender - Filters, tuning, better similarity

#### Refined Preprocessing (Filters)

```python
# Filter out books with no description or fewer than 50 characters
df_filtered = df_books[df_books['description'].str.len() > 50].copy()

# Add a weight to the categories to make them "stronger" in the vector space
# By repeating the category name, we increase its TF-IDF score
df_filtered['enhanced_features'] = df_filtered['description'] + (" " + df_filtered['categories']) * 3
```

#### Tunning

```python
# Tuning: use n-grams (1, 2) to capture context like "romantic novel" or "social satire"
tfidf_tuned = TfidfVectorizer(
    stop_words='english', 
    ngram_range=(1, 2), 
    min_df=2,    # Ignore terms that appear in less than 2 books
    max_features=5000
)
tfidf_matrix_tuned = tfidf_tuned.fit_transform(df_filtered['enhanced_features'])
```

#### Better Similarity

```python
from sklearn.preprocessing import normalize

# Normalizing the matrix improves similarity results by focusing on 
# the relative distribution of words rather than the length of the summary
tfidf_matrix_norm = normalize(tfidf_matrix_tuned)
cosine_sim_tuned = linear_kernel(tfidf_matrix_norm, tfidf_matrix_norm)
```

#### Try the optimized model

```python
# Re-run clustering on the tuned, normalized matrix
num_clusters = 10
kmeans_tuned = KMeans(n_clusters=num_clusters, n_init=10, random_state=42)
df_filtered['cluster'] = kmeans_tuned.fit_predict(tfidf_matrix_norm)

# Now your recommendations will be:
# 1. Filtered for quality
# 2. Tuned for contextual phrases (n-grams)
# 3. Normalized for better mathematical comparison
```

```python
def get_recommendations_refined(title, cosine_sim=cosine_sim_tuned):
    # Usamos df_filtered porque es donde ahora vive la columna 'cluster'
    indices = pd.Series(df_filtered.index, index=df_filtered['title']).drop_duplicates()
    
    if title not in indices:
        return f"'{title}' dataset not found."
    
    idx = indices[title]
    target_cluster = df_filtered.loc[idx, 'cluster']
    
    cluster_indices = df_filtered[df_filtered['cluster'] == target_cluster].index
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = [score for score in sim_scores if score[0] in cluster_indices and score[0] != idx]
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[:3]
    
    book_indices = [i[0] for i in sim_scores]
    return df_filtered['title'].iloc[book_indices]
    
print("Better results:")
print(get_recommendations_refined("Pride and Prejudice"))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.cluster import KMeans

# 1. NLP Vectorization (Already have data, so let's rebuild the matrix)
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_books['combined_features'])

# 2. Clustering (Group books into 10 thematic clusters)
num_clusters = 10
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df_books['cluster'] = kmeans.fit_predict(tfidf_matrix)

# 3. Similarity Matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# 4. Refined Recommendation Function
def get_recommendations(title, cosine_sim=cosine_sim):
    indices = pd.Series(df_books.index, index=df_books['title']).drop_duplicates()
    
    if title not in indices:
        return "Book not found."
    
    idx = indices[title]
    target_cluster = df_books.iloc[idx]['cluster']
    
    # Filter by cluster, then calculate similarity
    cluster_indices = df_books[df_books['cluster'] == target_cluster].index
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Filter for books in the same cluster, excluding the book itself
    sim_scores = [score for score in sim_scores if score[0] in cluster_indices and score[0] != idx]
    
    # Fallback to general search if cluster is too small
    if len(sim_scores) < 2:
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = [score for score in sim_scores if score[0] != idx]
        
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[:3]
    
    book_indices = [i[0] for i in sim_scores]
    return df_books['title'].iloc[book_indices]

# Final test
print(get_recommendations("Pride and Prejudice"))

```

The model has successfully identified thematic clusters. By using K-Means, you forced the computer to find "books that feel like this one" based on the vector space, not just common words. The clustering acted as a filter that pushed these highly-relevant Austen/Classic titles to the top.
